```python
import pandas as pd

df2_soundtracks = pd.read_csv('data/set1.csv')
df2_soundtracks['Max_emotion'] = df2_soundtracks[['Anger', 'Fear', 'Happy','Sad', 'Tender']].idxmax(axis="columns")
df2_soundtracks

```

```python
# df2_soundtracks
# 3D plot Valence vs. Energy vs. Tension, colour by emotion
# ax2 = df1_users_songs.plot.scatter(x='valenceValue',y='arousalValue')
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
z = df2_soundtracks[['Tension']]
x = df2_soundtracks[['Valence']]
y = df2_soundtracks[[ 'Energy']]
 
# Creating figure
fig = plt.figure(figsize = (7, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot #'Anger', 'Fear', 'Happy','Sad', 'Tender'
colors = {'Anger':'red', 'Fear':'green', 'Sad':'blue', 'Happy':'yellow', 'Tender':'purple'}

ax.scatter3D(x, y, z, color = df2_soundtracks['Max_emotion'].map(colors))
ax.set_xlabel('Valence')
ax.set_ylabel('Energy')
ax.set_zlabel('Tension')

plt.title("Soundtrack annotations")
#ax.legend(colors) 
# show plot
plt.show()
```

```python
# Print class distributions
df2_soundtracks.Max_emotion.value_counts().plot(kind='bar', title='Distribution of emotion categories in Soundtracks as Stimuli dataset', xlabel="Emotion categories", ylabel="Frequency")
```
